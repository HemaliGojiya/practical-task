DISPLAYAND CONVER INTO CAMEL CAS

```python
data={
    "car items": 
    [{"name": "car cover", "price": 1200, "stock": 5},
    {"name": "car perfume", "price": 250, "stock": 10},
    {"name": "car washer", "price": 500, "stock": 3}],

    "school items": 
    [{"name": "notebook", "price": 50, "stock": 100},
    {"name": "pen", "price": 10, "stock": 200}],
    
    "home items": 
    [{"name": "bucket", "price": 150, "stock": 20},
    {"name": "mop", "price": 300, "stock": 15}],
    
    "office items": 
    [{"name": "printer paper", "price": 400, "stock": 50},
    {"name": "stapler", "price": 80, "stock": 25}]
}

def to_camel_case(name):
    return name.split()[0] + ''.join(word.capitalize() for word in name.split()[1:])

for items in data.values():
    for item in items:
        print(to_camel_case(item["name"]), item["price"])
```

TOTAL VALUES

```python
data={
    "car items": 
    [{"name": "car cover", "price": 1200, "stock": 5},
    {"name": "car perfume", "price": 250, "stock": 10},
    {"name": "car washer", "price": 500, "stock": 3}],

    "school items": 
    [{"name": "notebook", "price": 50, "stock": 100},
    {"name": "pen", "price": 10, "stock": 200}],
    
    "home items": 
    [{"name": "bucket", "price": 150, "stock": 20},
    {"name": "mop", "price": 300, "stock": 15}],
    
    "office items": 
    [{"name": "printer paper", "price": 400, "stock": 50},
    {"name": "stapler", "price": 80, "stock": 25}]

}
def total_valuation(data):
    total = 0
    for items in data.values():
        for item in items:
            total += item["price"] * item["stock"]
    return total

print("Total Value:", total_valuation(data))
```

SEARCHING

```python
data={
    "car items": 
    [{"name": "car cover", "price": 1200, "stock": 5},
    {"name": "car perfume", "price": 250, "stock": 10},
    {"name": "car washer", "price": 500, "stock": 3}],

    "school items": 
    [{"name": "notebook", "price": 50, "stock": 100},
    {"name": "pen", "price": 10, "stock": 200}],
    
    "home items": 
    [{"name": "bucket", "price": 150, "stock": 20},
    {"name": "mop", "price": 300, "stock": 15}],
    
    "office items": 
    [{"name": "printer paper", "price": 400, "stock": 50},
    {"name": "stapler", "price": 80, "stock": 25}]
}
def search_product(data, keyword):
    for category, items in data.items():
        for item in items:
            if keyword.lower() in item["name"].lower():
                print("Found in:", category)
                print("Price:", item["price"], "Stock:", item["stock"])

search_product(data, "car")
```

HIEGHST VALUES

```python
data={
    "car items": 
    [{"name": "car cover", "price": 1200, "stock": 5},
    {"name": "car perfume", "price": 250, "stock": 10},
    {"name": "car washer", "price": 500, "stock": 3}],

    "school items": 
    [{"name": "notebook", "price": 50, "stock": 100},
    {"name": "pen", "price": 10, "stock": 200}],
    
    "home items": 
    [{"name": "bucket", "price": 150, "stock": 20},
    {"name": "mop", "price": 300, "stock": 15}],
    
    "office items": 
    [{"name": "printer paper", "price": 400, "stock": 50},
    {"name": "stapler", "price": 80, "stock": 25}]
}
def highest_value_category(data):
    max_value = 0
    max_category = ""

    for category, items in data.items():
        total = sum(item["price"] * item["stock"] for item in items)
        if total > max_value:
            max_value = total
            max_category = category

    print("Highest Category:", max_category, "Value:", max_value)

highest_value_category(data)
```
